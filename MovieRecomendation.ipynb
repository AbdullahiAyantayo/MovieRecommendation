```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
movies = pd.read_csv('/content/drive/MyDrive/movies.csv')
```

```python
movies.head()
```

```python
movies.fillna('')
```

```python
combo = movies['genres'] + ' ' + movies['keywords'] + ' ' + movies['original_title'] + ' ' + movies['overview'] + ' ' + movies['tagline'] + ' ' + movies['title'] + ' ' + movies['cast'] + ' ' +  movies['director']
```

```python
combo.head()
```

```python
combo.shape
```

```python
print(combo)
```

```python
combo = combo.fillna('')
```

```python
feature_extraction = TfidfVectorizer()
```

```python
values = feature_extraction.fit_transform(combo)
```

```python
print(values)
```

```python
similiarity = cosine_similarity(values)
```

```python
similiarity.shape
```

```python
print(similiarity)
```
